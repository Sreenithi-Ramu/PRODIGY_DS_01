```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "population_data.csv"

# Load and inspect data
data = pd.read_csv(file_path, skiprows=2)  # Adjust based on dataset format
data.rename(columns=lambda x: x.strip(), inplace=True)  # Clean column names
print("Column Names:", data.columns)
print("First Few Rows:\n", data.head())

# Filter for a specific year (adjust column name if necessary)
year = 2020
if "Year" not in data.columns:
    print("Year column not found. Available columns:", data.columns)
else:
    filtered_data = data[data["Year"] == year]

    # Bar Chart for Top Countries by Population
    filtered_data = filtered_data.sort_values(by="Population", ascending=False).head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x="Country Name", y="Population", data=filtered_data, palette="viridis")
    plt.title(f"Top 10 Countries by Population in {year}", fontsize=16)
    plt.xlabel("Country", fontsize=12)
    plt.ylabel("Population", fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Select relevant columns: Country Name and 2020 (or any year of your choice)
data_selected = data[['Country Name', '2020']]

# Clean any NaN values or filter out unwanted rows
data_selected = data_selected.dropna()

# Optional: Select specific countries (if needed)
countries = ['Aruba', 'Afghanistan', 'Angola']
data_selected = data_selected[data_selected['Country Name'].isin(countries)]

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(data_selected['Country Name'], data_selected['2020'], color='skyblue')

# Adding labels and title
plt.xlabel('Country')
plt.ylabel('Population in 2020')
plt.title('Population of Selected Countries in 2020')
plt.xticks(rotation=45)  # Rotate country names for better readability

# Show the plot
plt.tight_layout()
plt.show()
```
